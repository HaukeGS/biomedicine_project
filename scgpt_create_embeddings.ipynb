```python
model_dir = Path("/kaggle/input/scgpt_whole_human/pytorch/default/4")
covid_data_path = "/kaggle/input/covid19-split0/split_0.h5ad"
covid_data_small_path = "/kaggle/input/covid19-split-small-0/split_0.h5ad"
time_data_path = "/kaggle/input/pbmcs-covid-19-patients-over-time/may24_covid19.h5ad"
sample_data_path = "/kaggle/input/scgpt-tutorial-example/human_pancreas_norm_complexBatch.h5ad"
```

```python
# run the notebook multiple times with different splits
```

```python
adata = sc.read_h5ad(covid_data_small_path)

gene_col = "gene_symbol"
cell_type_key = "celltype"
batch_key = "tech"
N_HVG = 1800

adata.var[gene_col] = adata.var.index.values
# org_adata = adata.copy()
```

```python
print(f"Adata is:\n {adata}")
print("##########################################################")
print(f"Adata.var:\n {adata.var.head()}")
print("##########################################################")
print(f"Adata.var.index.values:\n {adata.var.index.values}")
print("##########################################################")
print(f"Adata.obs:\n {adata.obs.head()}")
print("##########################################################")
# print(f"Adata.obsm: {adata.obsm.head()}")
print(f"Adata.X:\n {adata.X}")
```

```python
# preprocess
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
# highly variable genes
sc.pp.highly_variable_genes(adata, n_top_genes=N_HVG, flavor='seurat_v3')
adata = adata[:, adata.var['highly_variable']]
```

```python
embed_adata = scg.tasks.embed_data(
    adata,
    model_dir,
    gene_col=gene_col,
    batch_size=64,
)
```
